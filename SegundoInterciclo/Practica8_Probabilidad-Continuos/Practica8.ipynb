######  Se determina distribucion de probabilidad continua cuando representan un número que pertenece al rango de los números reales. 

 

*   **Distribución de Pareto :** es una distribución de probabilidad continua con dos parámetros, que tiene aplicación en disciplinas como la sociología, la geofísica y la economía.

Tiene dos aplicaciones principales:

1. Modelar la distribución de ingresos.
2. Modelar la distribución de las poblaciones de la ciudad.


  La Distribución de Pareto esta dada por la formula:
  
  **F(x) = 1 – (k/x) α**

  Donde: 
  x es la variable aleatoria

  k es el límite inferior de los datos

  α es el parámetro de forma




```python
# Graficando Pareto
k = 2.3 # parametro de forma.
pareto = stats.pareto(k)
x = np.linspace(pareto.ppf(0.01),
                pareto.ppf(0.99), 100)
pareto_pdf = pareto.pdf(x) # Función de Probabilidad
plt.plot(x, pareto_pdf)
plt.title('Distribución de Pareto')
plt.ylabel('Probabilidad')
plt.xlabel('Valores')
plt.show()
```

```python
# histograma
aleatorios = pareto.rvs(1000) # genera aleatorios
cuenta, cajas, ignorar = plt.hist(aleatorios, 20)
plt.ylabel('frequencia')
plt.xlabel('valores')
plt.title('Histograma de Pareto')
plt.show()
```

*   Distribucion de t de Student: es un modelo teórico utilizado para aproximar el momento de primer orden de una población normalmente distribuida cuando el tamaño de la muestra es pequeño y se desconoce la desviación típica.

Su formula esta dada por:

image.png


  Donde:

  X = Media poblacional

  u = Media muestral

  s = Desviacion Estandar o tipica

  raiz cuadrado n = Numero de datos 

```python
df = 50
t = stats.t(df)
x = np.linspace(t.ppf(0.01),
                t.ppf(0.99), 100)
t_fp = t.pdf(x) # Función de Probabilidad
plt.plot(x, t_fp)
plt.title('Distribución t de Student')
plt.ylabel('Probabilidad')
plt.xlabel('Valores')
plt.show()
```

```python
# histograma
aleatorios = t.rvs(1000) # genera aleatorios
cuenta, cajas, ignorar = plt.hist(aleatorios, 20)
plt.ylabel('frequencia')
plt.xlabel('valores')
plt.title('Histograma t de Student')
plt.show()
```

**Aplicacion**

La distribución t se utiliza cuando: 

* Queremos estimar la media de una población normalmente distribuida a partir de una muestra pequeña. 
* Tamaño de la muestra es inferior a 30 elementos, es decir, n < 30. 

A partir de 30 observaciones, la distribución t se parece mucho a la distribución normal y, por tanto, utilizaremos la distribución normal.

* No se conoce la desviación típica o estándar de una población y tiene que ser estimada a partir de las observaciones de la muestra.

  * **Chi Cuadrado:** es un método de prueba de hipótesis ampliamente utilizado para contar datos. Método de prueba de hipótesis común basado en la distribución.Su hipótesis básica H0 es: no hay diferencia entre la frecuencia observada y la frecuencia esperada.

  La formula correspondiente es: 

  image.png

  Donde:
  * La variable x>= 0 y el parámetro  es un número entero positivo
  * El grado de libertad es un número entero positivo 


```python
# Graficando Chi cuadrado

# parametro de forma.
df = 34
chi2 = stats.chi2(df)
x = np.linspace(chi2.ppf(0.01),
                chi2.ppf(0.99), 100)

# Función de Probabilidad
chi2_fp = chi2.pdf(x)
plt.plot(x, chi2_fp)
plt.title('Distribución Chi cuadrado')
plt.ylabel('Probabilidad')
plt.xlabel('Valores')
plt.show()
```

```python
# Generación de valores aleatorios
aleatorios = chi2.rvs(1000)
cuenta, cajas, ignorar = plt.hist(aleatorios, 20)
plt.ylabel('frequencia')
plt.xlabel('valores')
plt.title('Histograma Chi cuadrado')
plt.show()
```

**Ejemplo:** Se seleccionan al azar dos clases de una escuela secundaria y se investiga su actitud hacia las artes liberales. Como resultado, 37 personas en la clase A están de acuerdo y 27 en desacuerdo; 39 personas en la clase B están de acuerdo y 21 personas se oponen. ¿Hay una diferencia significativa en la actitud (α = .05)?

Puede dibujar uno 2*2 2∗2Línea de * ∗Tabla bidireccional de columna (tablas de referencias cruzadas 2 * 2)



```python
from  scipy.stats import chi2_contingency
import numpy as np
kf_data = np.array([[37,27], [39,21]])
kf = chi2_contingency(kf_data)
print('chisq-statistic=%.4f, p-value=%.4f, df=%i expected_frep=%s'%kf)
# chisq-statistic=0.4054, p-value=0.5243, df=1 expected_frep=[[39.22580645 24.77419355][36.77419355 23.22580645]]
```

  * **Beta:** es una familia de distribuciones continuas de probabilidad definidas en el intervalo (0, 1) parametrizada por dos parámetros positivos de forma, denotados por *(alpha)  y (beta)* , que aparecen como exponentes de la variable aleatoria y controlan la forma de la distribución.

  La formula correspondiente es: 

  image.png

  Donde:
  * Los parámetros p y q son números reales positivos
  * La variable x satisface la condicion 0<= x <=1
  * B(p,q) es la funcion beta


```python
# Graficando Beta

# Parametros de forma
a, b = 2.3, 0.6
beta = stats.beta(a, b)
x = np.linspace(beta.ppf(0.01),
                beta.ppf(0.99), 100)

# Función de Probabilidad
beta_fp = beta.pdf(x)
plt.plot(x, beta_fp)
plt.title('Distribución Beta')
plt.ylabel('Probabilidad')
plt.xlabel('Valores')
plt.show()
```

```python
# Generación de valores aleatorios
aleatorios = beta.rvs(1000)
cuenta, cajas, ignorar = plt.hist(aleatorios, 20)
plt.ylabel('frequencia')
plt.xlabel('valores')
plt.title('Histograma Beta')
plt.show()
```

  * **Log-Normal:** Una variable X sigue una distribución log-normal si los logaritmos neperianos de sus valores están normalmente distribuidos, es decir, si la variable Y=log X es N(u,o) es decir si: 

  image.png

  En ella observamos que:

  * Depende de los parámetros m y σ.

  * Su media y varianza son:

  image.png



```python
# Graficando Log-Normal
sigma = 0.6
lognormal = stats.lognorm(sigma)
x = np.linspace(lognormal.ppf(0.01),
                lognormal.ppf(0.99), 100)

# Función de Probabilidad
lognorm_fp = lognormal.pdf(x)
plt.plot(x, lognorm_fp)
plt.title('Distribución Log-normal')
plt.ylabel('Probabilidad')
plt.xlabel('Valores')
plt.show()
```

```python
# Generación de valores aleatorios
aleatorios = lognormal.rvs(1000)
cuenta, cajas, ignorar = plt.hist(aleatorios, 20)
plt.ylabel('frequencia')
plt.xlabel('valores')
plt.title('Histograma Log-normal')
plt.show()
```
