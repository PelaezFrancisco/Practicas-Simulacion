#Datos Fifa 

```python
#Librerias
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_fifa= pd.read_csv("https://github.com/PelaezFrancisco/Practicas-Simulacion/raw/main/Practica2/fifa_datos.csv")
data_fifa=data_fifa.drop(columns = ['Unnamed: 0'])
data_fifa.head()
```

Describe(), permite encontrar las estadisticas de los datos

```python
data_fifa.describe()
```

Info(), imprime un resumen de los datos como el tiempo, indice, columnas, nulos, uso de memoria

```python
data_fifa.info()
```

Permite encontrar la correlacion que existe entre columnas(variables)

```python
data_fifa.corr(method ='pearson')
```

```python
plt.subplots(figsize=(15,10)) 
sns.heatmap(data_fifa.corr())
```
